```python
z = (1-(2*fo)-(2*Bi*fo))
print(z)
```

```python
p=1
T0_loop=(fo*((2*T1)**p + x)) + (y*(T0**p))

T1_loop=(fo*((2*T0)**p + x)) + (y*(T1**p))

T2_loop=(fo*((2*T1)**p + x)) + (y*(T2**p))

T3_loop=(fo*((2*T2)**p + x)) + (y*(T3**p))

T4_loop=(fo*((2*T3)**p + x)) + (y*(T4**p))

#Para o valor de T5 

T5_loop=(0.750*((T4**p) + (Bi*T_infinito) + (x/2))) + (z*(T5**p))

print(T0_loop)
print(T1_loop)
print(T2_loop)
print(T3_loop)
print(T4_loop)
print(T5_loop)
```
