A partir disso, vamos isolar o termo de Ti e calcular o valor procurado.

Declarando as variáveis abaixo:

```python
T_vizinhanca = 300
T_infinito = 300
h = 20
T_externa = 400
E = 0.8
k = 0.7
L = 0.05
alfa = 5.67*(10**-8) # constante de Stefan-Boltzmann.

```

Isolando o termo Ti na equação acima temos: 

```python
T_interna = ((h*(T_externa - T_infinito) + E*alfa*(T_externa**4 - T_vizinhanca**4))*(L/k) ) + T_externa

T_interna_formatado = '{:.2f}'.format(T_interna)

print('A temperatura da superfície interna é',T_interna_formatado,'K')
```

Note que não utilizamos o valor de área.

E se fosse pedido o fluxo de valor por área:

image.png

```python
q_cond = (k*(T_interna - T_externa))/ L 

q_cond_formatado = '{:.2f}'.format(q_cond)

print('O fluxo de valor por área é',q_cond_formatado,'W/m2')
```
